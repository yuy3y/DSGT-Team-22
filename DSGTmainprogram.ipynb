````python
# Open data
import pandas as pd
import tensorflow as tf
import numpy
from tensorflow import keras

matches = pd.read_csv("DSGT.csv")
matches
````

````python
# Process Data
matches = matches[["HomeTeam", "AwayTeam", "FTR"]].copy()

matches
````

````python
# This jsut creates an array of all instnces of all teams
teams = []
for i in matches["HomeTeam"]:
  teams.append(i)
for i in matches["AwayTeam"]:
  teams.append(i)

keys = list(set(teams))
values = []

for i in range(len(keys)):
  values.append(i)

print(values)
print(keys)
````

````python
teamDict = {}

for i in range(len(keys)):
  teamDict[keys[i]] = values[i]

print(teamDict)

for i in range (len(matches["HomeTeam"])):
  matches["HomeTeam"][i] = teamDict.get(matches["HomeTeam"][i])
for i in range (len(matches["AwayTeam"])):
  matches["AwayTeam"][i] = teamDict.get(matches["AwayTeam"][i])

matches
````

````python
FTRDict = {
    'H':0,
    'D':1,
    'A':2       
           }

for i in range (len(matches["FTR"])):
  matches["FTR"][i] = FTRDict.get(matches["FTR"][i])

matches
````

````python
matches["HomeTeam"] = tf.convert_to_tensor(matches["HomeTeam"], dtype=tf.int16)
matches["AwayTeam"] = tf.convert_to_tensor(matches["AwayTeam"], dtype=tf.int16)
matches["FTR"] = tf.convert_to_tensor(matches["FTR"], dtype=tf.int16)

matches.dtypes
````

````python
# Split into testing and training
split = 0.9

training_set = matches[:int(len(matches)*split)]
test_set = matches[int(len(matches)*split):]

train_features = training_set[["HomeTeam", "AwayTeam"]].copy()
train_labels = training_set[["FTR"]].copy()

test_features = test_set[["HomeTeam", "AwayTeam"]].copy()
test_labels = test_set[["FTR"]].copy()

train_labels.dtypes
````

````python
# Machine Learning
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(2,)),
    keras.layers.Dense(8, activation="relu"),
    keras.layers.Dense(8, activation="relu"),
    keras.layers.Dense(8, activation="relu"),
    keras.layers.Dense(3, activation="softmax"),
])

model.compile(optimizer="adam", loss="sparse_categorical_crossentropy", metrics=["accuracy"])\

model.fit(train_features, train_labels["FTR"], epochs=11)

````

````python
# Testing accuracy
test_loss, test_accuracy = model.evaluate(test_features, test_labels["FTR"])
print("Tested accuracy: ",test_accuracy*100,"%")

prediction = model.predict(test_features)
````
